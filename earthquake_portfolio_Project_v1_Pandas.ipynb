```python
import pandas as pd
import numpy as np
```

```python
earthquake = pd.read_csv(r"D:/Google Drive/Sheets/data.csv/data.csv")
```

```python
earthquake.info()
```

```python
# reviewing the data
earthquake.head()
```

```python
# Dropping duplicates in the dataset
earthquake = earthquake.drop_duplicates()
```

```python
earthquake
```

```python
# Convert the Date column to string
earthquake = earthquake.astype({'Date': 'str'})

# Convert the Date column to datetime64[ns] format
earthquake['Date'] = pd.to_datetime(earthquake['Date'], format='%d/%m/%Y')

earthquake.info()
```

```python
earthquake
```

```python


# Clean the Latitude and Longitude columns by removing non-numeric characters except for digits, decimal points, and negative signs
earthquake['Latitude'] = earthquake['Latitude'].apply(lambda x: '-' + x.replace(' S', '') if ' S' in x else x.replace(' N', ''))
earthquake['Longitude'] = earthquake['Longitude'].apply(lambda x: '-' + x.replace(' W', '') if ' W' in x else x.replace(' E', ''))

# Remove non-numeric characters from the Latitude and Longitude columns 
earthquake['Latitude'] = earthquake['Latitude'].str.replace(r'[^0-9.-]', '', regex=True)
earthquake['Longitude'] = earthquake['Longitude'].str.replace(r'[^0-9.-]', '', regex=True)


earthquake.head(351)
```

```python
# Drop the 'Map' column from the dataframe
earthquake = earthquake.drop(columns=['Map'])
earthquake
```

```python
# Rename the columns in the dataframe for cleaner column names

earthquake.rename(columns={'Mode': 'Detection_Type'}, inplace=True)
earthquake.rename(columns={'Time (utc)': 'Time_in_UTC'}, inplace=True)
earthquake.rename(columns={'Depth (km)': 'Depth_in_km'}, inplace=True)
```

```python

# Function to ensure only one decimal point in a string
def ensure_one_decimal(val):
    parts = val.split('.')
    if len(parts) > 2:
        return parts[0] + '.' + ''.join(parts[1:])
    return val

# Function to clean the values by removing non-numeric characters except for digits, decimal points, and negative signs
def clean_value(val):
    val = ''.join(c for c in val if c.isdigit() or c == '.' or c == '-')
    return ensure_one_decimal(val)

# Apply the cleaning function to the columns
earthquake['Latitude'] = earthquake['Latitude'].apply(clean_value)
earthquake['Longitude'] = earthquake['Longitude'].apply(clean_value)

# Convert to float
earthquake['Latitude'] = earthquake['Latitude'].astype(float)
earthquake['Longitude'] = earthquake['Longitude'].astype(float)
```
